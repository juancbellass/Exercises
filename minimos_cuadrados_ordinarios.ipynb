```
import numpy as np
import scipy as sc

import matplotlib.pyplot as plt 
```

```
from sklearn.datasets import load_boston
```

```
#Cargamos la libreria
boston = load_boston()

print(boston.DESCR)
```

Fórmula de mínimoz cuadrados ordinarios (MCO): $\beta = (X^{T}X)^{-1}X^{T}Y$

```
X = np.array(boston.data[:, 5])
Y = np.array(boston.target)

plt.scatter(X, Y, alpha=0.3)
plt.show()
```

```
X.shape
```

```
#Anhadimos columna de 1s para terminos independientes.
X1 = np.array([np.ones(506), X]).T

B = np.linalg.inv(X1.T @ X1) @ X1.T @ Y
```

```
# Visualizamos nuestra matriz Beta
B 
```

```
#vemos el resultado final
plt.scatter(X, Y, alpha=0.3)
plt.plot([4,9], [B[0] + B[1] * 4, B[0] + B[1] * 9], c="red")
plt.show()
```
