```
import numpy as np
import scipy as sc

import matplotlib.pyplot as plt
```

```
func = lambda th: np.sin(1 / 2 * th[0] ** 2 - 1 / 4 * th[1] ** 2 + 3) * np.cos(2 * th[0] + 1 - np.e ** th[1]) #funcion de costo

res = 100 #resolucion

_X = np.linspace(-2, 2, res)
_Y = np.linspace(-2, 2, res)

_Z = np.zeros([res, res]) #creo una matriz de ceros para cargar los valores de la funcion evaluadas en (_X, _Y)

for ix, x in enumerate(_X):
  for iy, y in enumerate(_Y):
    _Z[iy, ix] = func([x, y]) #voy evaluando y cargando a la matriz _Z

#plt.contour(_X, _Y, _Z)
plt.contourf(_X, _Y, _Z, 100)
plt.colorbar()

Theta = np.random.rand(2) * 4 - 2  #evaluo la funcion genrerando un vector random (x, y) ajustado a un intervalo (-2, 2)

_T = np.copy(Theta) #copio la variable Theta en otra para ir manipulandola

h = 0.001 #intervalo diferencial
lr = 0.01 #ratio de aprendizaje

plt.plot(Theta[0], Theta[1], "o", c="white") #grafico elpunto de inicio

grad = np.zeros(2) #definimos primeramente al gradiente como un vector (0,0)

for _ in range(1000): #hacemos 1000 iteraciones
  for it, th in enumerate(Theta):
    
    _T = np.copy(Theta)
  
    _T[it] = _T[it] + h 
  
    deriv = (func(_T) - func(Theta)) / h
  
    grad[it] = deriv
  
  Theta = Theta - lr * grad 
  
  if _ % 100 == 0:
    plt.plot(Theta[0], Theta[1], ".", c="red")

plt.plot(Theta[0], Theta[1], "o", c="green")
plt.show
```
